```
# Import semua library
# Pandas untuk membaca dokumen csv
# Numpy untuk perhitungan matriks dan bilangan 
# Matplotlib untuk mengeplot grafik
# Seaborn untuk visualisasi grafik di python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
# Membaca dataframe
df = pd.read_csv("datapenjualan.csv")
df
```

```
df.head(10)
```

```
# Statistik deskriptif
df.describe()
```

```
# Menghitung jumlah lokasi 
df['location'].value_counts()
```

```
# Menghitung subcategory
df['subcategory'].value_counts()
```

```
# Mengubah jumlah yang terjual ke bentuk interger 
df['quantity_sold'] = df['quantity_sold'].astype(int)
```

```
#menghitung jumlah produk yang dijual untuk tiap subkategori
produk_paling_banyak_dijual = df[['subcategory', 'shop_name']].groupby(['subcategory']).count().sort_values(by='shop_name',ascending=False).reset_index()
produk_paling_banyak_dijual.head()
```

```
# Menggunakan barplot dengan fungsi sns
sns.barplot(x='subcategory',y='shop_name',data=produk_paling_banyak_dijual[:10],color="green")
plt.xlabel("Sub-kategori")
plt.ylabel("Jumlah Nama Toko")
plt.title("Jumlah produk berdasarkan sub-kategori")
plt.xticks(rotation='vertical')
plt.show()
```

```
# Mengubah tipe data jumlah view sebagai integer
df['jumlah_view'] = df['jumlah_view'].astype(int)
```

```
df.describe()
```

```
# Melihat data subcategory dengan jumlah yang lihat
# Namun data masih belum dijumlahkan sehingga perlu fungsi group by  
subcategory_jumlah_view = df[['subcategory', 'jumlah_view']]
subcategory_jumlah_view.head(10)
```

```
# Menghitung jumlah view per subcategory dengan fungsi group by
jumlah_view_per_subcategory = subcategory_jumlah_view.groupby(['subcategory']).sum()
jumlah_view_per_subcategory.head()
```

```
# Mengurutkan jumlah view setiap subcategory
subcategory_jumlah_view = jumlah_view_per_subcategory.sort_values(by='jumlah_view',ascending=False).reset_index()
subcategory_jumlah_view.head()
```

```
# Ploting sub kategori dengan jumlah produk yang dilihat diurutkan dari yang terbanyak
sns.barplot(x='subcategory',y='jumlah_view',data=subcategory_jumlah_view[:10],color="black")
plt.xlabel("Sub-kategori")
plt.ylabel("Jumlah Produk yang Dilihat")
plt.title("Sub-kategori yang paling banyak dilihat")
plt.xticks(rotation='vertical')
plt.show()
```

```
# Menampilkan subcategory dengan quantity sold namun belum dihitung jumlahnya 
# Jadi harus pakai fungsi group by
subcategory_quantity_sold = df[['subcategory', 'quantity_sold']]
subcategory_quantity_sold.head()
```

```
# Menghitung jumlah produk yang terjual berdasarkan subcategory
# Perhitungan dilakukan dengan fungsi group by 
quantity_sold_per_subcategory = subcategory_quantity_sold.groupby(['subcategory']).sum()
quantity_sold_per_subcategory.head()
```

```
# Mengurutkan jumlah pembelian produk 
subcategory_quantity_sold = quantity_sold_per_subcategory.sort_values(by='quantity_sold',ascending=False).reset_index()
subcategory_quantity_sold.head()
```

```
# Ploting data diurutkan dari yang paling banyak dibeli
sns.barplot(x='subcategory',y='quantity_sold',data=subcategory_quantity_sold[:10],color="red")
plt.xlabel("Sub-kategori")
plt.ylabel("Jumlah Terjual")
plt.title("Sub-kategori yang paling banyak terjual")
plt.xticks(rotation='vertical')
plt.show()
```

```
# conversion rates marketing = jumlah barang terjual / jumlah orang yang melihat * 100
# semakin tinggi conversion ratesnya maka semakin mudah produk tersebut terjual
# data masih dalam bentuk float
df['conversion_rates'] = (df.quantity_sold / df.jumlah_view)*100
df['conversion_rates'].head()
```

```
# Mengisi data yang kosong/ NaN dengan nilai 0
df['conversion_rates'] = df['conversion_rates'].fillna(0)
df['conversion_rates'].head()
```

```
df['conversion_rates'] = df['conversion_rates'].replace([np.inf, -np.inf], np.nan).dropna(axis=0)
```

```
title_conversion_rates = df[['title', 'conversion_rates']]
title_conversion_rates.head()
```

```
#conversion_rates_new = conversion_rates.sort_values(by='conversion_rates',ascending=False).reset_index()
#conversion_rates_new.head()
conversion_rates_new = title_conversion_rates.sort_values(by='conversion_rates',ascending=False)
conversion_rates_new.head(20)
```

```
# Jadi prodik teratas memiliki potensi penjualan produk terbesar 
sns.barplot(x='conversion_rates',y='title',data=conversion_rates_new[:10],color="pink")
plt.xlabel("Conversion Rates")
plt.ylabel("Nama Produk")
plt.title("Produk dengan conversion rates terbesar")
plt.xticks(rotation='vertical')
plt.show()
```
